```python
data = []
for line in input_str.strip().split('\n'):
    line_split = line.split(' | ')
    line_split[0] = line_split[0].split(' ')
    line_split[1] = line_split[1].split(' ')
    data.append(line_split)
```

```python
def find_bigger_string(small_string, pattern):
    for big_string in pattern:
        if len(big_string) == len(small_string) + 1:
            is_match = True
            for x in small_string:
                if x not in big_string:
                    is_match = False
            if is_match:
                for x in big_string:
                    if x not in small_string:
                        return x
```

```python
def find_smaller_string(big_string, pattern):
    for small_string in pattern:
        if len(small_string) == len(big_string) - 1:
            is_match = True
            for x in small_string:
                if x not in big_string:
                    is_match = False
            if is_match:
                for x in big_string:
                    if x not in small_string:
                        return x
```

```python
def get_output_num(pattern, output):
    num_to_pattern = {}
    for x in pattern:
        if len(x) in digits_to_num:
            num_to_pattern[digits_to_num[len(x)]] = x

    top = find_bigger_string(num_to_pattern[1], pattern)
    bottom = find_bigger_string(num_to_pattern[4] + top, pattern)
    num_to_pattern[9] = num_to_pattern[4] + top + bottom
    bottom_left = find_smaller_string(''.join([x for x in num_to_pattern[8] if x not in (top, bottom)]), pattern)
    top_left = find_bigger_string(num_to_pattern[7] + bottom + bottom_left, pattern)
    num_to_pattern[0] = num_to_pattern[7] + bottom + bottom_left + top_left
    num_to_pattern[3] = ''.join([x for x in num_to_pattern[8] if x not in (top_left, bottom_left)])
    bottom_right = find_smaller_string(num_to_pattern[3] + bottom_left, [p for p in pattern if ''.join(sorted(p)) != ''.join(sorted(num_to_pattern[3]))])
    num_to_pattern[2] = ''.join([x for x in num_to_pattern[3] + bottom_left if x not in (bottom_right)])

    not_found = []
    for x in pattern:
        found = False
        for y in num_to_pattern.values():
            if ''.join(sorted(x)) == ''.join(sorted(y)):
                found = True
        if found == False:
            not_found.append(x)

    if len(not_found[0]) < len(not_found[1]):
        num_to_pattern[5] = not_found[0]
        num_to_pattern[6] = not_found[1]
    else:
        num_to_pattern[6] = not_found[0]
        num_to_pattern[5] = not_found[1]

    output_num = ''
    for x in output:
        for num, y in num_to_pattern.items():
            if ''.join(sorted(x)) == ''.join(sorted(y)):
                output_num += str(num)

    output_num = int(output_num)
    
    return output_num
```
