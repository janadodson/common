```python
input_image = np.array([[int(x == '#') for x in list(line)] for line in input_str.strip().split('\n')])
```

```python
def expand_image(input_image, expand_num):
    image = deepcopy(input_image)

    image = np.append(image, [[expand_num] * image.shape[1]], axis=0)
    image = np.append([[expand_num] * image.shape[1]], image, axis=0)

    image = np.append(image, [[expand_num]] * image.shape[0], axis=1)
    image = np.append([[expand_num]] * image.shape[0], image, axis=1)

    return image
```

```python
def get_output_pixel(input_image, x, y):
    grp = input_image[x-1:x+2, y-1:y+2].flatten()
    return int(''.join(grp.astype(str)), 2)
```

```python
def convert_image(input_image, expand_num):
    image = deepcopy(input_image)
    
    for i in range(2):
        image = expand_image(image, expand_num)
        
    output_image = np.full(image.shape, 1-expand_num)
    
    for x in range(1, image.shape[0]-1):
        for y in range(1, image.shape[1]-1):
            i = get_output_pixel(image, x, y)
            output_image[x, y] = algo[i]
    
    return output_image
```

```python
output_image = deepcopy(input_image)
```

```python
for i in range(50):
    print(i, output_image.shape)
    if i % 2 == 0:
        expand_num = 0
    else:
        expand_num = 1
    output_image = convert_image(output_image, expand_num)
```
