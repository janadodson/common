```python
data = []
for line in input_str.strip().split('\n'):
    line_split = list(line)
    data.append(line_split)
```

```python
char_map = {
    '[': ']',
    '(': ')',
    '{': '}',
    '<': '>'
}
```

```python
def find_corrupt(line):
    pts = {
        ')': 3,
        ']': 57,
        '}': 1197,
        '>': 25137
    }
    
    openers = []
    for i in line:
        if i in char_map.keys():
            openers.append(i)
        else:
            last_opener = openers.pop()
            if char_map[last_opener] != i:
                return pts[i]
    return 0
```

```python
score = 0
keep_data = []
for line in data:
    temp_score = find_corrupt(line)
    score += temp_score
    if temp_score == 0:
        keep_data.append(line)
```

```python
def find_incomplete(line):
    pts = {
        ')': 1,
        ']': 2,
        '}': 3,
        '>': 4
    }
    
    openers = []
    for i in line:
        if i in char_map.keys():
            openers.append(i)
        else:
            openers.pop()
            
    closers = []
    for i in reversed(openers):
        closers.append(char_map[i])
        
    score = 0
    for x in closers:
        score = score * 5 + pts[x]
        
    return score
```

```python
scores = []
for line in keep_data:
    scores.append(find_incomplete(line))
```
