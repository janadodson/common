```python
def find_intersection(coords_1, coords_2):
    (x_min_1, x_max_1), (y_min_1, y_max_1), (z_min_1, z_max_1) = coords_1
    (x_min_2, x_max_2), (y_min_2, y_max_2), (z_min_2, z_max_2) = coords_2

    x_min, x_max = max(x_min_1, x_min_2), min(x_max_1, x_max_2)
    y_min, y_max = max(y_min_1, y_min_2), min(y_max_1, y_max_2)
    z_min, z_max = max(z_min_1, z_min_2), min(z_max_1, z_max_2)

    if x_min <= x_max and y_min <= y_max and z_min <= z_max:
        return (x_min, x_max), (y_min, y_max), (z_min, z_max)
```

```python
def cube_volume(coords):
    (x_min, x_max), (y_min, y_max), (z_min, z_max) = coords
    return (x_max - x_min + 1) * (y_max - y_min + 1) * (z_max - z_min + 1)
```
