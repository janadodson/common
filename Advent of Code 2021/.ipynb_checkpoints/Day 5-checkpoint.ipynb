```python
def part_1(data):
    counts = {}
    for p1, p2 in data:
        if p1[0] == p2[0] or p1[1] == p2[1]:
            for x in range(min(p1[0], p2[0]), max(p1[0], p2[0]) + 1):
                for y in range(min(p1[1], p2[1]), max(p1[1], p2[1]) + 1):
                    if (x, y) in counts:
                        counts[(x, y)] += 1
                    else:
                        counts[(x, y)] = 1

    c = 0
    for p, count in counts.items():
        if count > 1:
            c += 1

    return c
```

```python
part_1(data)
```

```python
def get_list(x1, x2):
    if x2 >= x1:
        return range(x1, x2 + 1)
    else:
        return range(x1, x2 - 1, -1)
```

```python
def add_count(x, y):
    if (x, y) not in counts:
        counts[(x, y)] = 0
    counts[(x, y)] += 1
```

```python
counts = {}
for p1, p2 in data:
    if p1[0] == p2[0]:
        for y in get_list(p1[1], p2[1]):
            add_count(p1[0], y)
    elif p1[1] == p2[1]:
        for x in get_list(p1[0], p2[0]):
            add_count(x, p1[1])
    else:
        for x, y in zip(get_list(p1[0], p2[0]), get_list(p1[1], p2[1])):
            add_count(x, y)
```
