```python
dots = [(int(dot.split(',')[0]), int(dot.split(',')[1])) for dot in input_str.split('\n\n')[0].strip().split('\n')]
```

```python
folds = []
for line in input_str.split('\n\n')[1].strip().split('\n'):
    line_split = line.split(' ')[-1].split('=')
    folds.append((line_split[0], int(line_split[1])))
```

```python
def fold_y(dots, y_fold):
    new_dots = set()
    for x, y in dots:
        if y < y_fold:
            new_dots.add((x, y))
        if y > y_fold:
            new_dots.add((x, y_fold - (y - y_fold)))
    return new_dots

def fold_x(dots, x_fold):
    new_dots = set()
    for x, y in dots:
        if x < x_fold:
            new_dots.add((x, y))
        if x > x_fold:
            new_dots.add((x_fold - (x - x_fold), y))
    return new_dots
```

```python
def fold(dots, axis, line):
    if axis == 'x':
        return fold_x(dots, line)
    if axis == 'y':
        return fold_y(dots, line)
```

```python
len(fold(dots, *folds[0]))
```

```python
for f in folds:
    dots = fold(dots, *f)
```

```python
max_x = 0
max_y = 0
for x, y in dots:
    max_x = max(max_x, x)
    max_y = max(max_y, y)
```

```python
for x in range(max_x + 1):
    for y in range(max_y + 1):
        if (x, y) in dots:
            a[x, y] = '#'
        else:
            a[x, y] = '.'
```
