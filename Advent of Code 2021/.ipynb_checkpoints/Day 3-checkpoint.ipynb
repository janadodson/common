```python
def process_input(input_str):
    data = []
    for line in input_str.strip().split('\n'):
        line_split = line.strip().split()
        line_split[0] = line_split[0]
        line_split[1] = int(line_split[1])
        data.append(line_split)
    return data
```

```python
def part_1(data):
    h, d = 0, 0
    for direction, distance in data:
        if direction == 'forward':
            h += distance
        elif direction == 'down':
            d += distance
        elif direction == 'up':
            d -= distance
    return h * d
```

```python
def part_2(data):
    a, h, d = 0, 0, 0
    for direction, distance in data:
        if direction == 'forward':
            h += distance
            d += a * distance
        elif direction == 'down':
            a += distance
        elif direction == 'up':
            a -= distance
    return h * d
```

```python
data = process_input(input_str)
```

```python
part_1(data)
```

```python
part_2(data)
```
